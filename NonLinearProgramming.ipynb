### Non-linear Optimization - Exercises

```python
import gurobipy as gp
```

```python
m = gp.Model()
```

```python
x1 = m.addVar(vtype=gp.GRB.CONTINUOUS, name='x1')
x2 = m.addVar(vtype=gp.GRB.CONTINUOUS, name='x2')
```

```python
m.setObjective(x1**2 + 2*x2**2)
```

```python
m.optimize()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to minimize
def f(x1, x2):
    return x1**2 + 2*x2**2

# Define the gradient of the function
def gradient(x1, x2):
    df_dx1 = 2 * x1
    df_dx2 = 4 * x2
    return np.array([df_dx1, df_dx2])

# Gradient Descent Parameters
learning_rate = 0.1
num_iterations = 20

# Initial guess
x1, x2 = 8, 8  # You can choose any initial values

# Lists to store the values for plotting
x1_values = [x1]
x2_values = [x2]
f_values = [f(x1, x2)]

# Gradient Descent Iteration
for i in range(num_iterations):
    grad = gradient(x1, x2)
    x1 -= learning_rate * grad[0]
    x2 -= learning_rate * grad[1]
    x1_values.append(x1)
    x2_values.append(x2)
    f_values.append(f(x1, x2))

# Plot the contour plot of the function
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, 10, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = X1**2 + 2*X2**2

plt.figure(figsize=(8, 6))
contour = plt.contour(X1, X2, Z, levels=20, cmap='viridis')
plt.clabel(contour, inline=True, fontsize=10)

# Plot the iterations
plt.plot(x1_values, x2_values, marker='o', linestyle='-', color='red', markersize=4)

# Add labels and a color bar
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar()

# Show the plot
plt.title('Gradient Descent Steps on $f(x_1, x_2) = x_1^2 + 2x_2^2$')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function to minimize
def f(x1, x2):
    return x1**2 + 2*x2**2

# Define the gradient of the function
def gradient(x1, x2):
    df_dx1 = 2 * x1
    df_dx2 = 4 * x2
    return np.array([df_dx1, df_dx2])

# Gradient Descent Parameters
learning_rate = 0.02
num_iterations = 80  # Set to 5 for the first five iterations

# Initial distant point
x1, x2 = 10, 10  # Choose a distant initial point

# Lists to store the values for plotting
x1_values = [x1]
x2_values = [x2]
f_values = [f(x1, x2)]

# Gradient Descent Iteration
for i in range(num_iterations):
    grad = gradient(x1, x2)
    x1 -= learning_rate * grad[0]
    x2 -= learning_rate * grad[1]
    x1_values.append(x1)
    x2_values.append(x2)
    f_values.append(f(x1, x2))

# Create a 3D plot of the function
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, 10, 400)
X1, X2 = np.meshgrid(x1, x2)
Z = X1**2 + 2*X2**2

plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')

# Plot the surface
contour = ax.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.5)

# Plot the iterations
ax.scatter(x1_values, x2_values, f_values, c='red', marker='o', s=20)

# Plot the gradient vectors
scale_factor = 0.  # Adjust this for arrow length
for i in range(len(x1_values)):
    grad = gradient(x1_values[i], x2_values[i])
    ax.quiver(x1_values[i], x2_values[i], f_values[i], 
              grad[0], grad[1], 0, color='red', length=scale_factor, normalize=True)

# Add a red line connecting the gradient steps
ax.plot(x1_values, x2_values, f_values, color='red', linewidth=2, marker='o', markersize=1, label='Gradient Descent Path')

# Add labels
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1, x2)')
plt.title('Gradient Descent Steps')

# Show the plot
plt.grid(True)
plt.legend()
plt.show()
```
